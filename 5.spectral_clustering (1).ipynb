```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import SpectralClustering
clustering = SpectralClustering(n_clusters=10, affinity='rbf')
y_labels = clustering.fit_predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_labels)
plt.show()
```

```python
supervised=pd.DataFrame(dataset)
supervised['cluster_group']=y_labels
supervised
```
